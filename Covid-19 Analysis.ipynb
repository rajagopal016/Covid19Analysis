10. Write a Python program to create a plot (lines) of total deaths, confirmed, recovered and active cases Country wise where deaths greater than 150.

```python
df10 = covid_data
df10 = df10.groupby(['Country/Region'])['Confirmed','Deaths','Recovered'].sum().reset_index()
df10['Active'] = df10['Confirmed'] - (df10['Deaths'] - df10['Recovered'])
df10=df10[df10['Deaths'] > 150]
df10
plt.figure(figsize=(15, 5))
plt.plot(df10['Country/Region'], df10['Deaths'], color='red')
plt.plot(df10['Country/Region'], df10['Confirmed'], color='Blue')
plt.plot(df10['Country/Region'], df10['Recovered'], color='Green')
plt.plot(df10['Country/Region'], df10['Active'], color='Yellow')

plt.title('Total Deaths(>150), Confirmed, Recovered and Active Cases by Country')
plt.show()
```

11. Write a Python program to visualize the state/province wise death cases of Novel Coronavirus (COVID-19) in USA.

```python
df11 = covid_data
df11 = df11[df11['Country/Region'] == 'US'].reset_index(drop=1)
df11 = df11[df11['Deaths'] > 0]

fig = plt.figure(figsize=(20,10))
ax = fig.add_axes([0,0,1,1])
ax.bar(df11['Province/State'],df11['Deaths'])
ax.set_xticklabels(df11['Province/State'], rotation=90, ha='right')
plt.title("State/Province wise death cases of Novel Coronavirus (COVID-19) in USA", fontsize=20 )
plt.show()
```

12. Write a Python program to visualize the state/province wise Active cases of Novel Coronavirus (COVID-19) in USA

```python
df12 = covid_data
df12["Active"] = df12["Confirmed"] - df12["Recovered"] - df12['Deaths']
df12 = df12[df12["Country/Region"] == "US"].reset_index()
df12 = df12.groupby(["Province/State"])["Province/State","Confirmed", "Active", "Deaths", "Recovered"].sum().reset_index()
df12 = df12[df12["Active"] > 0]
#df12
#plt.plot(  )

fig = plt.figure(figsize=(20,10))

ax = fig.add_axes([0,0,1,1])
ax.bar(df12['Province/State'],df12['Active'])
ax.set_xticklabels(df12['Province/State'], rotation=90, ha='right')
plt.title("State/Province wise Active cases of Novel Coronavirus (COVID-19) in USA", fontsize=20 )
```

13. Write a Python program to visualize Worldwide Confirmed Novel Coronavirus (COVID-19) cases over time.
